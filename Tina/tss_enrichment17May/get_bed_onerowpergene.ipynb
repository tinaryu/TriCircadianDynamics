```python
import pandas as pd
```

```python

chip = 'H3K27ac' #this should be H3K9ac or H327ac
df = pd.read_excel('onerowperGene.xlsx',sheet_name=chip)
df
```

```python
bed = pd.DataFrame(df['chr'])
```

```python
bed['chromStart'] = df['chromStart']
bed['chromEnd'] = df['chromEnd']
bed['name'] = df['gene']
bed = bed.assign(score=0)
bed['strand'] = df['Strand']
```

```python
bed
```

```python
bed.to_csv(chip+'_beds.bed',sep='\t',header=None,index=False)
```
