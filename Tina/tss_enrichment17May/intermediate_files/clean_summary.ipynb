```python
#this script takes the coordinates in one line and breaks it down to separate columns
import pandas as pd
import re
```

```python
df = pd.read_excel('summary.xlsx')
```

```python
df
```

```python
chrnum = [re.search(r"chr[^_]+",x).group() for x in df['peak']]
chr_start_end = [re.search(r"\d*_\d*$",x).group() for x in df['peak']]
chrSE = pd.DataFrame(chr_start_end)[0].str.split("_",expand=True)
chrSE
```

```python
df['chr'] = chrnum
df['chromStart'] = chrSE[0]
df['chromEnd'] = chrSE[1]
df
```

```python
newdf = df[['peak','chr', 'chromStart', 'chromEnd', 'CV_amplitude', 'CV_phase', 'CV_mean', 'CV_pvalue', 'CV_qvalue', 'CV_fdr_BH', 'GF_amplitude', 'GF_phase', 'GF_mean', 'GF_pvalue','GF_qvalue', 'GF_fdr_BH']]
```

```python
newdf
```

```python
filt = ~(newdf['chr'] == 'chrUn')
filt.value_counts()
```

```python
newdf = newdf.loc[filt]
newdf
```

```python
newdf.to_csv('H3K9ac.csv',index=False)
```
