```python
import pandas as pd
```

This code takes summary_edited.xlsx and filters out the genes with distance greater than 10kb
The resulting data is saved as a .csv file


```python
df = pd.read_excel('summary_edited.xlsx',sheet_name='H3K9ac')
```

```python
df
```

```python
filt = df['Distance'] < 10000
filt.value_counts()
```

```python
df = df.loc[filt]
```

```python
df
```

```python
df.to_csv('H3K9ac_distfilt.csv',index=False)
```
