## The purpose of this file is to creating a pie chart, showing the amount of genes that are oscillating in CV and GF groups

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# sns.color_palette()[8]
# sns.color_palette()[9]

# Set Seaborn style
sns.set_style('whitegrid')

# Data
total_genes = 23288
cv = [11051, total_genes - 11051]
gf = [6122, total_genes - 6122]

color1 = sns.color_palette()[0]
color2 =sns.color_palette()[7]

labels = ['Oscillating', 'Non-Oscillating']
colors = [color1, color2]  # Seaborn default blue and orange

# Custom function for labels to include count and percentage
def autopct_format(pct, all_vals):
    absolute = int(round(pct * sum(all_vals) / 100))
    return f'{absolute:,} ({pct:.1f}%)'

# Create figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# CV Pie Chart
axes[0].pie(cv, labels=labels, autopct=lambda pct: autopct_format(pct, cv), 
            startangle=90, colors=colors, wedgeprops={'edgecolor': 'white'})
axes[0].set_title('CV', fontsize=20)

# GF Pie Chart
axes[1].pie(gf, labels=labels, autopct=lambda pct: autopct_format(pct, gf), 
            startangle=90, colors=colors, wedgeprops={'edgecolor': 'white'})
axes[1].set_title('GF', fontsize=20)

# Ensure equal aspect ratio for perfect circles
for ax in axes:
    ax.axis('equal')

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style
sns.set_style('whitegrid')

# Data
total_genes = 23288
cv = [11051, total_genes - 11051]
gf = [6122, total_genes - 6122]

# Define colors
color1 = sns.color_palette("Paired")[4]
color2 = sns.color_palette()[7]
labels = ['Oscillating', 'Non-Oscillating']
colors = [color1, color2]  # Seaborn default blue and orange

# Custom function for labels to include count and percentage
def autopct_format(pct, all_vals):
    absolute = int(round(pct * sum(all_vals) / 100))
    return f'{absolute:,} ({pct:.1f}%)'

# CV Pie Chart
fig_cv, ax_cv = plt.subplots(figsize=(5, 5))
ax_cv.pie(cv, labels=labels, autopct=lambda pct: autopct_format(pct, cv), 
          startangle=90, colors=colors, wedgeprops={'edgecolor': 'white'})
ax_cv.set_title('CV', fontsize=20)
ax_cv.axis('equal')  # Ensure equal aspect ratio

plt.show()
```

```python
# GF Pie Chart

# Define colors
color1 = sns.color_palette("Paired")[2]
color2 = sns.color_palette()[7]
labels = ['Oscillating', 'Non-Oscillating']
colors = [color1, color2]  # Seaborn default blue and orange

sns.color_palette("Set2")
fig_gf, ax_gf = plt.subplots(figsize=(5, 5))
ax_gf.pie(gf, labels=labels, autopct=lambda pct: autopct_format(pct, gf), 
          startangle=90, colors=colors, wedgeprops={'edgecolor': 'white'})
ax_gf.set_title('GF', fontsize=20)
ax_gf.axis('equal')  # Ensure equal aspect ratio

plt.show()
```
