```python
import pandas as pd
```

```python
df = pd.read_csv('tss_regions_mm10_1000.bed',sep='\t',header=None)
df
```

```python
random_sample = df.sample(n=500)
random_sample = random_sample.sort_index()

random_sample
```

```python
h = pd.read_csv('hdac3_uniq_ID.txt')
f = pd.read_csv('nfil3_uniq_ID.txt')
r = pd.read_csv('nr1d1_uniq_ID.txt')
h
```

```python
h_filt = df[3].isin(h['CycID'])
f_filt = df[3].isin(f['CycID'])
r_filt = df[3].isin(r['CycID'])
```

```python
hdac3_genes = df.loc[h_filt]
nfil3_genes = df.loc[f_filt]
nr1d1_genes = df.loc[r_filt]
nfil3_genes
```

```python
hdac3_genes.to_csv('1000/hdac3_beds.bed',sep='\t',header=None,index=False)
nfil3_genes.to_csv('1000/nfil3_beds.bed',sep='\t',header=None,index=False)
nr1d1_genes.to_csv('1000/nr1d1_beds.bed',sep='\t',header=None,index=False)
random_sample.to_csv('1000/random_beds.bed',sep='\t',header=None,index=False)
```
