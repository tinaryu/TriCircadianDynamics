```python
import pandas as pd
from matplotlib import pyplot as plt

#takes summary_distfilt.xlsx excel file and makes a onerowperGene.xlsx
#only containing the rows with distance closest to the tss coordinate
```

```python
df_K9 = pd.read_excel('summary_edited.xlsx',sheet_name='H3K9ac')
df_K9
```

```python
df_K27 = pd.read_excel('summary_edited.xlsx',sheet_name='H3K27ac')
df_K27.rename(columns={'Name_b':'gene'}, inplace=True)
df_K27
```

```python
df_comb = [df_K9, df_K27]
```

```python
#cast type on data obj
for df in df_comb:
    df['gene'] = df['gene'].astype(str)
```

```python
gene_grp_K9 = df_K9.groupby('gene')
gene_grp_K9.get_group('Mrpl15')
```

```python
gene_grp_K27 = df_K27.groupby('gene')
gene_grp_K27.get_group('Mrpl15')
```

```python
means_min_K9 = gene_grp_K9['Distance'].transform('min')
means_min_K27 = gene_grp_K27['Distance'].transform('min')
```

```python
means_min_K9
```

```python
filt = df_K9['Distance'] == means_min_K9
result_K9 = df_K9[filt]
result_K9
```

```python
result_K27 = df_K27[df_K27['Distance'] == means_min_K27]
result_K27
```

```python
with pd.ExcelWriter('onerowperGene_no_distfilt.xlsx') as writer:
    result_K9.to_excel(writer, sheet_name='H3K9ac', index = False)
    result_K27.to_excel(writer, sheet_name='H3K27ac',index=False)
```
