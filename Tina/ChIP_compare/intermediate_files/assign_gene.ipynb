```python
#import pandas as pd
import pyranges as pr
```

assign gene to each peak, based on closest distance

```python
coordinates = pr.read_bed('H3K9ac_coords.bed')
genes = pr.read_bed('tss_regions_mm10.bed')

closest_genes = coordinates.nearest(genes)
type(closest_genes)
```

```python
df = closest_genes.df
df
```

```python
df.sort_values(by=['Name'], inplace=True)
df
```

```python
df.to_csv('H3K9_nearest.csv',index=False)
```
