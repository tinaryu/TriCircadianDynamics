```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.cluster.hierarchy import dendrogram, linkage

# Function to normalize a single column
def normalize_col(column):
    return (column - column.min()) / (column.max() - column.min())

# Function to calculate differences and normalize them
def calculate_and_normalize_differences(df_wt, df_ko):
    # Calculate the amplitude difference according to the specified formula
    raw_amp_diff = df_wt['meta2d_rAMP'] - df_ko['meta2d_rAMP']
    corrected_amp_diff = raw_amp_diff - np.log(np.abs(raw_amp_diff))
    amp_diff_normalized = normalize_col(corrected_amp_diff)
    
    # Directly calculate and normalize phase differences
    raw_phase_diff = df_wt['meta2d_phase'] - df_ko['meta2d_phase']
    phase_diff_normalized = normalize_col(raw_phase_diff)
    
    return pd.DataFrame({
        'CycID': df_wt['CycID'],
        'amp_diff': amp_diff_normalized,
        'phase_diff': phase_diff_normalized
    })

# Gene groups with their corresponding files
gene_groups = [
    ("meta2d_hdac3_WT.txt", "meta2d_hdac3_KO.txt", "hdac3_difference.txt"),
    ("meta2d_nr1d1_WT.txt", "meta2d_nr1d1_KO.txt", "nr1d1_difference.txt"),
    ("meta2d_nfil3_WT.txt", "meta2d_nfil3_KO.txt", "nfil3_difference.txt"),
]

# Initialize list for storing processed DataFrames
data_frames = []

# Process each gene group and save differences
for wt_file, ko_file, output_file in gene_groups:
    df_wt = pd.read_csv(wt_file, sep='\t')
    df_ko = pd.read_csv(ko_file, sep='\t')
    difference_df = calculate_and_normalize_differences(df_wt, df_ko)
    data_frames.append(difference_df)
    # Save the differences to a text file
    difference_df.to_csv(output_file, sep='\t', index=False)

# Calculate distance matrix
distance_matrix = np.zeros((len(data_frames), len(data_frames)))

for i in range(len(data_frames)):
    for j in range(i+1, len(data_frames)):
        dist = np.mean(np.sqrt(np.square(data_frames[i]['amp_diff'] - data_frames[j]['amp_diff']) + 
                               np.square(data_frames[i]['phase_diff'] - data_frames[j]['phase_diff'])))
        distance_matrix[i, j] = dist
        distance_matrix[j, i] = dist

# Labels for plotting
labels = ["hdac3_difference", "nr1d1_difference", "nfil3_difference"]

# Plotting the heatmap
sns.heatmap(distance_matrix, annot=True, fmt=".2f", xticklabels=labels, yticklabels=labels)
plt.title('Distance Matrix of Normalized Differences')
plt.show()

# Performing hierarchical clustering and plotting the dendrogram
Z = linkage(distance_matrix, 'average')
plt.figure(figsize=(6, 4))
dendrogram(Z, labels=labels, leaf_rotation=90, leaf_font_size=12)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Gene Group')
plt.ylabel('Distance')
plt.show()
```

```python
import pandas as pd

# Paths to the files
path_hdac3 = 'hdac3_difference.txt'
path_nfil3 = 'nfil3_difference.txt'
path_nr1d1 = 'nr1d1_difference.txt'

# Load the data files
df_hdac3 = pd.read_csv(path_hdac3, sep='\t')
df_nfil3 = pd.read_csv(path_nfil3, sep='\t')
df_nr1d1 = pd.read_csv(path_nr1d1, sep='\t')


```

```python
#Step 1: Define a function to convert continuous data to binary based on threshold
def to_binary(df, thresholds):
    binary_df = pd.DataFrame()
    for col in df.columns[1:]:  # Skipping the 'CycID' column
        binary_df[col] = (df[col] > thresholds[col]).astype(int)
    return binary_df
```

```python
# Concatenate the dataframes to calculate global thresholds
#Step 2: Calculate the median thresholds and convert data
all_data = pd.concat([df_hdac3.iloc[:, 1:], df_nfil3.iloc[:, 1:], df_nr1d1.iloc[:, 1:]])

# Calculate thresholds
thresholds = all_data.median()

# Convert data to binary based on thresholds
binary_hdac3 = to_binary(df_hdac3, thresholds)
binary_nfil3 = to_binary(df_nfil3, thresholds)
binary_nr1d1 = to_binary(df_nr1d1, thresholds)
```

```python
### Step 3: Define a function to calculate Jaccard index and compute similarities
def jaccard_similarity(binary_df1, binary_df2):
    intersection = (binary_df1 & binary_df2).sum().sum()  # Sum of intersection counts
    union = (binary_df1 | binary_df2).sum().sum()  # Sum of union counts
    return intersection / union

# Calculate Jaccard Similarity Index between each pair
jaccard_hdac3_nfil3 = jaccard_similarity(binary_hdac3, binary_nfil3)
jaccard_hdac3_nr1d1 = jaccard_similarity(binary_hdac3, binary_nr1d1)
jaccard_nfil3_nr1d1 = jaccard_similarity(binary_nfil3, binary_nr1d1)

jaccard_hdac3_nfil3, jaccard_hdac3_nr1d1, jaccard_nfil3_nr1d1
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
import numpy as np


# Convert similarities to distances
dist_hdac3_nfil3 = 1 - jaccard_hdac3_nfil3
dist_hdac3_nr1d1 = 1 - jaccard_hdac3_nr1d1
dist_nfil3_nr1d1 = 1 - jaccard_nfil3_nr1d1

# Create a symmetric matrix of distances
distance_matrix = np.array([
    [0, dist_hdac3_nfil3, dist_hdac3_nr1d1],
    [dist_hdac3_nfil3, 0, dist_nfil3_nr1d1],
    [dist_hdac3_nr1d1, dist_nfil3_nr1d1, 0]
])

# Perform hierarchical clustering
Z = linkage(distance_matrix, 'single')

# Plot dendrogram
plt.figure(figsize=(5, 4))
dendrogram(Z, labels=["hdac3", "nfil3", "nr1d1"], leaf_rotation=90, leaf_font_size=12)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample')
plt.ylabel('Distance')
plt.show()
```

```python
import gseapy as gp
import pandas as pd
gene_list_nr1d1 = pd.read_csv("nr1d1_dep_geneID.txt",header=0, sep="\t")
glist_nr1d1 = gene_list_nr1d1.squeeze().str.strip().to_list()
# if you are only intrested in dataframe that enrichr returned, please set outdir=None
enr_nr1d1_MH = gp.enrichr(gene_list=gene_list_nr1d1, 
                 gene_sets=['MSigDB_Hallmark_2020'],
                 organism='mouse', # don't forget to set organism to the one you desired! e.g. Yeast
                 outdir=None, # don't write to disk
                )
```

```python
gene_list_hdac3 = pd.read_csv("hdac3_dep_geneID.txt",header=0, sep="\t")
glist_hdac3 = gene_list_hdac3.squeeze().str.strip().to_list()
# if you are only intrested in dataframe that enrichr returned, please set outdir=None
enr_hdac3_MH = gp.enrichr(gene_list=gene_list_hdac3, 
                 gene_sets=['MSigDB_Hallmark_2020'],
                 organism='mouse', # don't forget to set organism to the one you desired! e.g. Yeast
                 outdir=None, # don't write to disk
                )
```

```python
gene_list_nfil3 = pd.read_csv("nfil3_dep_geneID.txt",header=0, sep="\t")
glist_nfil3 = gene_list_nfil3.squeeze().str.strip().to_list()
# if you are only intrested in dataframe that enrichr returned, please set outdir=None
enr_nfil3_MH = gp.enrichr(gene_list=gene_list_nfil3, 
                 gene_sets=['MSigDB_Hallmark_2020'],
                 organism='mouse', # don't forget to set organism to the one you desired! e.g. Yeast
                 outdir=None, # don't write to disk
                )
```

```python
enr_nr1d1_MH_results = enr_nr1d1_MH.results
enr_hdac3_MH_results = enr_hdac3_MH.results
enr_nfil3_MH_results = enr_nfil3_MH.results
```

```python

from scipy.cluster.hierarchy import dendrogram, linkage
# Combine the top terms into one DataFrame
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
# select top 20 terms
top_nr1d1_MH_terms = enr_nr1d1_MH_results.nsmallest(20, 'Adjusted P-value')
top_hdac3_MH_terms = enr_hdac3_MH_results.nsmallest(20, 'Adjusted P-value')
top_nfil3_MH_terms = enr_nfil3_MH_results.nsmallest(20, 'Adjusted P-value')

top_nr1d1_MH_terms['GeneList'] = 'nr1d1'
top_hdac3_MH_terms['GeneList'] = 'hdac3'
top_nfil3_MH_terms['GeneList'] = 'nfil3'
combined_results_MH = pd.concat([top_nr1d1_MH_terms, top_hdac3_MH_terms, top_nfil3_MH_terms])

# Remove duplicates if necessary
combined_results_MH = combined_results_MH.drop_duplicates(subset=['Term', 'GeneList'])

combined_results_MH['Adjusted P-value'] = combined_results_MH['Adjusted P-value'].replace(0, 1e-300)

# Pivot the combined DataFrame to create a matrix for the heatmap
heatmap_data_MH = combined_results_MH.pivot(index='Term', columns='GeneList', values='Adjusted P-value')

# Fill NaN values with 1 before log transformation, indicating no enrichment
heatmap_data_MH = heatmap_data_MH.fillna(1)

# Convert the p-values to -log10(p-value) for visualization
# This will now not result in infinite values because there are no zeros
heatmap_data_log_MH = -np.log10(heatmap_data_MH)

row_linkage_MH = linkage(heatmap_data_log_MH, method='average')
col_linkage_MH = linkage(heatmap_data_log_MH.T, method='average')

# Create the clustermap with seaborn
clustergrid_MH = sns.clustermap(heatmap_data_log_MH, row_linkage=row_linkage_MH, col_linkage=col_linkage_MH,
                             cmap='YlOrRd', annot=None, linewidths=.5, figsize=(10, 8))

# Set the title with a horizontal position using `set_title`
clustergrid_MH.fig.suptitle('Top 20 Enriched Terms Across Gene Lists with Hierarchical Clustering', 
                         ha='right')  # `ha` is short for horizontalalignment

# Adjust the position of the main plot to make space for the title
plt.subplots_adjust(top=0.88)

# Show the plot
plt.show()
```
