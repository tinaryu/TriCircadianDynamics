```python
import pandas as pd

# Load the markers data from the uploaded CSV file
markers_df = pd.read_csv('markerList.csv')

# Group by cluster and select the top 20 markers by adjusted p-value
top20_markers = markers_df.groupby('cluster').apply(lambda x: x.nsmallest(20, 'p_val_adj')).reset_index(drop=True)

# Create a new DataFrame where each column represents a cluster with its top 20 genes
top20_genes_per_cluster = top20_markers.pivot(columns='cluster', values='gene')

# Save the result to a new CSV file
top20_genes_per_cluster.to_csv('top20_genes_per_cluster.csv', index=False)

```

```python
import pandas as pd

# Load the markers data from the uploaded CSV file
markers_df = pd.read_csv('markerList.csv')

# Group by cluster and select the top 20 markers by adjusted p-value
top20_markers = markers_df.groupby('cluster').apply(lambda x: x.nsmallest(20, 'p_val_adj')).reset_index(drop=True)

top20_genes = top20_markers[['gene', 'cluster']]

# Pivoting the table to have clusters as columns and gene names as values
top20_genes_per_cluster_clean = top20_genes.pivot_table(index=None, columns='cluster', aggfunc=lambda x: list(x)[:20])

# Flatten the list of genes into separate rows
top20_genes_per_cluster_clean = pd.DataFrame(top20_genes_per_cluster_clean.to_dict()).apply(pd.Series.explode).reset_index(drop=True)

# Save the result to a new CSV file
top20_genes_per_cluster_clean.to_csv('top20_genes_per_cluster_clean.csv', index=False)


```
